4.Create Loyalty Flag for Exisiting Customers

```python
# Use groupby and transform() functions with np.max argument to find max order number for each user
df_ords_prods_merge['max_order'] = df_ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# view/print updated data frame
df_ords_prods_merge.head(20)
```

```python
# loc function to make flag
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['max_order'] <= 40) & (df_ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# shows frequency of values in new loyalty flag column
df_ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# view updated data frame with new column
df_ords_prods_merge.head()
```

5.check the basic statistics of the product prices for each loyalty category

```python
# Use groupby and agg() to determine basic statisitcs of spending habits by loyalty group\n",
df_ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

Observation: The average price of a product purchased by a new customer is about 3 dollars more than a loyal customer and the average price of a product purchased a regular customer is about 2 dollars more than of a loyal customer

6.Create a spending flag based on avg price

```python
# transform() to determine the mean price of products purchased by each user
df_ords_prods_merge['mean_product_price'] = df_ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Create spending flag column using loc()
df_ords_prods_merge.loc[df_ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# frequency of new spending_flag column,
df_ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# View relevant columns for updated df
df_ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head()
```

7.Create order frequency flag based on median price based on days_since_prior_order column

```python
# Use groupby() and transform() to see median number of days between orders for each user
df_ords_prods_merge['median_days_between_orders'] = df_ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# View relevant columns for updated DF
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head()
```

```python
# Create order frequency flag based on given criteria using loc function
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['median_days_between_orders'] <= 20) & (df_ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# seeing value count of new order_frequency column in Df
df_ords_prods_merge['order_frequency'].value_counts(dropna = False)
```

```python
df_ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency']].head()
```
