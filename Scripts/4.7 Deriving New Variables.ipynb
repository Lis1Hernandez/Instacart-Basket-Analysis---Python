1. Create 'Price Label/Range' Column

```python
# Define Function - Price Labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# price range function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# most expensive price in the price-range
df['prices'].max()
```

```python
# using loc method for price range function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# now using loc method on the entirety of the ords_prods_merge DF to apply price range function
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df_ords_prods_merge.loc[(df_ords_prods_merge['prices'] <= 15) & (df_ords_prods_merge['prices'] > 5), 'price_range_loc'] = 'Mid-range product'
```

```python
df_ords_prods_merge.loc[df_ords_prods_merge['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

2. Create 'Busiest Day' Column

```python
# changing order_dow abbreviation to full orders_day_of_week
df_ords_prods_merge.rename(columns = {'order_dow' : 'orders_day_of_week'}, inplace = True)
```

```python
# finds the frequency 
df_ords_prods_merge['orders_day_of_week'].value_counts(dropna=False)
```

```python
# use for-loop and the orders_dow key from project brief
result_1 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result_1.append("Busiest day")
  elif value == 4:
    result_1.append("Least busy")
  else:
    result_1.append("Regularly busy")
```

```python
result_1
```

```python
#Create new column 'busiest_day' in ords_prods_merge to view results later
df_ords_prods_merge['busiest_day'] = result_1
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

3.  Change 'Busiest Day' Column to 'Busiest Days' column - this creates groups called 'Busiest Days', 'Regularly Busy' and 'Slowest Days'

```python
#Create 'busiest_days' column from 'orders_day_of_week' column to identify two busiest days + orders_dow key from project brief
result_2 = []

for value in df_ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value ==1:
        result_2.append("Busiest days")
    elif value == 3 or value== 4:
        result_2.append("Slowest days")
    else:
        result_2.append("Regularly busy")
```

```python
result_2
```

```python
df_ords_prods_merge['busiest_days'] = result_2
```

```python
df_ords_prods_merge['busiest_days'].value_counts(dropna = False)
```

4.  Create new column 'busiest_period_of_day' - identifies time periods 'Most orders','Average orders', and 'Fewest orders'

```python
df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
result_3 = []

for value in df_ords_prods_merge["order_hour_of_day"]:
    if value == 10 or value == 11 or value == 14 or value == 15 or value == 13 or value == 12 or value == 16 or value == 9:
        result_3.append("Most orders")
    elif value == 17 or value == 8 or value == 18 or value == 19 or value == 20 or value == 7 or value == 21 or value == 22:
        result_3.append("Average orders")
    else:
        result_3.append("Fewest orders")
```

```python
result_3
```

```python
# create new column 'busiest_period_of day' in ords_prods_merge DF
df_ords_prods_merge['busiest_period_of_day'] = result_3
```
